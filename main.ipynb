### Preprocessing

```python
import pandas as pd
```

```python
filmList = pd.read_csv('data/ml-20m/movies.csv')
filmList
```

```python
filmList['year'] = [x[-5:-1] for x in list(filmList['title'])]
filmList['title'] = [x[:-7].strip() for x in list(filmList['title'])]
filmList
```

```python
setToDelete = set(filmList['year'])
for i in range(1891, 2016):
    setToDelete.discard(str(i))
setToDelete = list(setToDelete)
```

```python
filmList['filter'] = filmList['year'].isin(setToDelete)
filmList = filmList.drop(filmList[filmList['filter']].index)
filmList['year'] = [int(x) for x in list(filmList['year'])]
```

```python
filmList2 = pd.read_csv('data/moviesList2/movies.csv')
filmList2
```

```python
filmList2 = filmList2.rename(columns={'name':'title'})
filmList2
```

```python
movies = filmList.merge(filmList2, on=['title', 'year'])
movies
```

```python
movies = movies.drop(columns=['filter', 'rating', 'released', 'score', 'votes', 'score', 'budget', 'gross', 'runtime'])
movies
```

```python
anotherList = pd.read_csv('data/moviesList1/filmtv_movies.csv')
anotherList
```

```python
anotherList = anotherList.drop(columns=['filmtv_id', 'duration', 'avg_vote', 'critics_vote', 'public_vote', 'description', 'notes', 'total_votes'])
anotherList
```

```python
movies = movies.merge(anotherList, on=['title', 'year'])
movies
```

```python
movies = movies.drop(columns=['genre_x', 'genre_y'])
movies = movies.drop(columns=['country_x', 'directors'])
movies = movies.rename(columns={'country_y':'country'})
movies
```

```python
f = open( 'item_index2entity_id.txt', 'w' )
for id, entity in movies.iterrows():
    f.write(str(entity['movieId']) + '\t' + str(id) + '\n')
```

```python
f = open( 'ids_with_names.txt', 'w' )
for id, entity in movies.iterrows():
    f.write(str(entity['movieId']) + '\t' + str(id) + '\t' + str(entity['title'] + '\n'))
```

```python
entity_counter = len(movies)
entity_dict = {}
stats = {
    'humor': {0: entity_counter, 1: entity_counter + 1, 2: entity_counter + 2, 3: entity_counter + 3, 4: entity_counter + 4},
    'rhythm': {0: entity_counter + 5, 1: entity_counter + 6, 2: entity_counter + 7, 3: entity_counter + 8, 4: entity_counter + 9, 5: entity_counter + 10},
    'effort': {0: entity_counter + 11, 1: entity_counter + 12, 2: entity_counter + 13, 3: entity_counter + 14, 4: entity_counter + 15, 5: entity_counter + 16},
    'tension': {0: entity_counter + 17, 1: entity_counter + 18, 2: entity_counter + 19, 3: entity_counter + 20, 4: entity_counter + 21, 5: entity_counter + 22},
    'erotism': {0: entity_counter + 23, 1: entity_counter + 24, 2: entity_counter + 25, 3: entity_counter + 26, 4: entity_counter + 27}
}
entity_counter += 28
```

```python
f = open( 'kg_final.txt', 'w' )
for index, row in movies.iterrows():
    #genres
    genreList = row['genres'].split('|')
    for genre in genreList:
        
        if genre == '(no genres listed)':
            continue
        
        if genre not in entity_dict.keys():
            entity_dict[genre] = entity_counter
            entity_counter += 1
        f.write(str(index) + '\t' + '0' + '\t' + str(entity_dict[genre]) + '\n')
    
    #year
    if str(row['year']) not in entity_dict.keys():
        entity_dict[str(row['year'])] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '1' + '\t' + str(entity_dict[str(row['year'])]) + '\n')
    
    #director
    if row['director'] not in entity_dict.keys():
        entity_dict[row['director']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '2' + '\t' + str(entity_dict[row['director']]) + '\n')
    
    #writer
    if row['writer'] not in entity_dict.keys():
        entity_dict[row['writer']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '3' + '\t' + str(entity_dict[row['writer']]) + '\n')
    
    #star #4
    if row['star'] not in entity_dict.keys():
        entity_dict[row['star']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '4' + '\t' + str(entity_dict[row['star']]) + '\n')
    
    #company #5
    if row['company'] not in entity_dict.keys():
        entity_dict[row['company']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '5' + '\t' + str(entity_dict[row['company']]) + '\n')
    
    #country #6
    if row['country'] not in entity_dict.keys():
        entity_dict[row['country']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '6' + '\t' + str(entity_dict[row['country']]) + '\n')
    
    #actors #7
    if type(row['actors']) != float:
        actorList = row['actors'].split(', ')
        for actor in actorList:
            if actor not in entity_dict.keys():
                entity_dict[actor] = entity_counter
                entity_counter += 1
            f.write(str(index) + '\t' + '7' + '\t' + str(entity_dict[actor]) + '\n')
    
    #humor 8
    f.write(str(index) + '\t' + '8' + '\t' + str(stats['humor'][row['humor']]) + '\n')
    
    #rhythm 9
    f.write(str(index) + '\t' + '9' + '\t' + str(stats['rhythm'][row['rhythm']]) + '\n')
    
    #effort 10
    f.write(str(index) + '\t' + '10' + '\t' + str(stats['effort'][row['effort']]) + '\n')
    
    #tension 11
    f.write(str(index) + '\t' + '11' + '\t' + str(stats['tension'][row['tension']]) + '\n')
    
    #erotism 12
    f.write(str(index) + '\t' + '12' + '\t' + str(stats['erotism'][row['erotism']]) + '\n')
    

#print(entity_dict)
```

### ratings

```python
rates = pd.read_csv('data/ml-20m/ratings.csv')
rates
```

```python
idMapper = {}
for index, row in movies.iterrows():
    idMapper[row['movieId']] = index
```

```python
movieSet = set(movies['movieId'])
f = open( 'ratings_final.txt', 'w' )

userSet = set()

for index, row in rates.iterrows():
    if row['movieId'] in movieSet:
        userSet.add(row['userId'] - 1)
    if row['movieId'] in movieSet:
        f.write(str(int(row['userId'] - 1)) + '\t' + str(int(idMapper[row['movieId']])) + '\t' + ('1' if row['rating'] >= 4. else '0') + '\n')
```
