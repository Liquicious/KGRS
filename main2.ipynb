```python
import pandas as pd
```

```python
filmList = pd.read_csv('data/ml-20m/movies.csv')
filmList
```

```python
filter = [False for _ in range(len(filmList))]
for index, row in filmList.iterrows():
    lst = row['title'].split()
    if len(lst) < 2:
        filter[index] = True
        continue
    
    if not lst[-1][1:-1].isdigit():
        filter[index] = True
        continue
print(filter.count(True))
```

```python
filmList['filter'] = filter
filmList
```

```python
filmList = filmList.drop(filmList[filmList['filter']].index)
filmList
```

```python
filmList['year'] = [x[-5:-1] for x in list(filmList['title'])]
filmList['title'] = [x[:-7].strip() for x in list(filmList['title'])]
filmList
```

```python
filmList = filmList.drop(columns='filter')
filmList
```

```python
info = pd.read_csv('data/moviesList3/rt_movie_info.csv')
info
```

```python
print(filmList[['title','year']].duplicated().sum())
filmList = filmList.drop_duplicates(subset=['title','year'])
filmList[['title','year']].duplicated().sum()
```

```python
info = info.rename(columns={'movie_name':'title'})
```

```python
info = info.drop(columns=['movie_link', 'movie_info', 'poster_image', 'movie_info', 'in_theaters_date', 'on_streaming_date', 'runtime_in_minutes', 'box_office', 'critics_consensus', 'tomatometer_status',
                          'tomatometer_rating', 'tomatometer_count', 'audience_status', 'audience_want_to_see_count', 'audience_rating', 'audience_count', 'audience_top_critics_count',
                          'audience_fresh_critics_count', 'audience_rotten_critics_count', 'audience_fresh_top_critics_count', 'audience_rotten_rotten_critics_count', 'tomatometer_avg_rating',
                          'audience_top_critics_avg_rating', 'score_sentiment', 'audience_avg_rating'])
info
```

```python
movies = filmList.merge(info, on='title')
movies
```

```python
movies.duplicated().sum()
```

```python
movies = movies.drop_duplicates()
movies.duplicated().sum()
```

```python
movies = movies.drop(columns='genre')
```

```python
movies = movies.reset_index()
movies
```

```python
movies = movies.drop(columns='index')
movies
```

```python
entity_counter = len(movies)
entity_dict = {}

f = open( 'kg_final.txt', 'w' )
for index, row in movies.iterrows():
    #genres
    genreList = row['genres'].split('|')
    for genre in genreList:
        
        if genre == '(no genres listed)':
            continue
        
        if genre not in entity_dict.keys():
            entity_dict[genre] = entity_counter
            entity_counter += 1
        f.write(str(index) + '\t' + '0' + '\t' + str(entity_dict[genre]) + '\n')
    
    #year
    if str(row['year']) not in entity_dict.keys():
        entity_dict[str(row['year'])] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '1' + '\t' + str(entity_dict[str(row['year'])]) + '\n')
    
    #rating
    if row['rating'] not in entity_dict.keys():
        entity_dict[row['rating']] = entity_counter
        entity_counter += 1
    f.write(str(index) + '\t' + '2' + '\t' + str(entity_dict[row['rating']]) + '\n')
    
    #director
    if type(row['directors']) != float:
        directorList = row['directors'].split(',')
        for director in directorList:
            if director not in entity_dict.keys():
                entity_dict[director] = entity_counter
                entity_counter += 1
            f.write(str(index) + '\t' + '3' + '\t' + str(entity_dict[director]) + '\n')
    
    #writer
    if type(row['writers']) != float:
        writerList = row['writers'].split(',')
        for writer in writerList:
            if writer not in entity_dict.keys():
                entity_dict[writer] = entity_counter
                entity_counter += 1
            f.write(str(index) + '\t' + '4' + '\t' + str(entity_dict[writer]) + '\n')
        
    #casts
    if type(row['casts']) != float:
        castsList = row['casts'].split(',')
        for casts in castsList:
            if casts not in entity_dict.keys():
                entity_dict[casts] = entity_counter
                entity_counter += 1
            f.write(str(index) + '\t' + '5' + '\t' + str(entity_dict[casts]) + '\n')
    
    #studio
    if type(row['studio_name']) != float:
        if row['studio_name'] not in entity_dict.keys():
            entity_dict[row['studio_name']] = entity_counter
            entity_counter += 1
        f.write(str(index) + '\t' + '6' + '\t' + str(entity_dict[row['studio_name']]) + '\n')
    
print(len(entity_dict))
```

```python
rates = pd.read_csv('data/ml-20m/ratings.csv')
rates
```

```python
idMapper = {}
for index, row in movies.iterrows():
    idMapper[row['movieId']] = index
```

```python
movieSet = set(movies['movieId'])
f = open( 'ratings_final.txt', 'w' )

userSet = set()

for index, row in rates.iterrows():
    if row['movieId'] in movieSet:
        userSet.add(row['userId'] - 1)
    if row['movieId'] in movieSet:
        f.write(str(int(row['userId'] - 1)) + '\t' + str(int(idMapper[row['movieId']])) + '\t' + ('1' if row['rating'] >= 4. else '0') + '\n')
```

```python
for i in range(max(rates['userId'])):
    if i not in userSet:
        print(i)
```
